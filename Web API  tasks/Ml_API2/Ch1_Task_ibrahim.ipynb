```python
import pandas as pd
import sklearn
import psycopg2 as pg
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC


engine = pg.connect("dbname='ML' user='postgres' host='127.0.0.1' port='5432' password='0000'")
df = pd.read_sql('SELECT * FROM public."Iris_API_iris_model"', con=engine)
df
```

```python
X = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = df['Species']

# Create training and testing vars, It’s usually around 80/20 or 70/30.
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.20, random_state=1)

# Now we’ll fit the model on the training data
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train, Y_train)

# Make predictions on validation dataset
predictions = model.predict(X_test)


# Pickle model 
pd.to_pickle(model,r'/home/ibrahim/Desktop/projects/Graduation-Project-Tasks/Web API  tasks/Ml_API/new_model.pickle')

# Unpickle model 
model = pd.read_pickle(r'/home/ibrahim/Desktop/projects/Graduation-Project-Tasks/Web API  tasks/Ml_API/new_model.pickle') 
# read a pickle pd.read_pickle('model.pkl')

```

```python
print("Test set score: {:.2f}".format(model.score(X_test, Y_test)))
```
